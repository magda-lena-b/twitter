### Wczytanie danych uzupełnionych w testach lokalizacji

```python
import pickle
with open('geo.pickle', 'rb') as f:
    geo = pickle.load(f)
```

Usuwam te, gdzie jest None lub spoza listy województw.

W przysłości - do poprawienia.

```python
geo.dropna()
for i in geo.index:
    if geo.state.loc[i] not in list(a.state):
        geo = geo.drop([i])
#type(list(a.jpt_nazwa_))
```

```python

geo.fav_cnt = pd.to_numeric(geo.fav_cnt)
geo.friends_cnt = pd.to_numeric(geo.friends_cnt)
geo.foll_cnt = pd.to_numeric(geo.foll_cnt)
```

```python
geo.head(1)
```

### Tworzenie agregatów do wykresów

Ponieważ teraz nie analizuję tekstu a użytkowników - tzreba distinct.

```python
geo=geo.iloc[:,1:9]
```

```python
geo=geo.drop(['fav_cnt'], axis=1).drop_duplicates()
```

```python
d2plot =geo.groupby('state')['foll_cnt'].agg(['size', 'mean'])
```

```python
d2plot['mean_friends']=geo.groupby('state')['friends_cnt'].mean()
d2plot
```

```python
d2plot = d2plot.reset_index()
```

### Łączenie z danymi geo

```python
d3plot = pd.merge(a, d2plot, on='state')
```

```python
d3plot
```

```python
print('Liczba użytkowników do analizy to' , d3plot['size'].agg('sum'))
```

```python
fig, ax = plt.subplots(1, figsize=(8, 8))
d3plot.plot(ax = ax, linewidth=0.5, edgecolor='0.4', column='size', cmap='Blues', legend=True, scheme='quantiles')
ax.set_title('Number of users in sample', fontdict={'fontsize': 20}, loc='left')
ax.get_legend().set_bbox_to_anchor((.2, .3))
ax.axis('off')
```

```python
fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(131)
d3plot.plot(ax=ax1, linewidth=0.5, edgecolor='0.4', column='size', cmap='Blues', legend=True, scheme='quantiles')
ax1.set_title('Number of users in sample', fontdict={'fontsize': 20}, loc='left')
ax1.get_legend().set_bbox_to_anchor((.4, .3))
ax1.axis('off')

ax2 = fig.add_subplot(132)
d3plot.plot(ax = ax2, linewidth=0.5, edgecolor='0.4', column='mean_friends', cmap='Reds', legend=True, scheme='quantiles')
ax2.set_title('Average number of friends', fontdict={'fontsize': 20}, loc='left')
ax2.get_legend().set_bbox_to_anchor((.4, .3))
ax2.axis('off')

ax1 = fig.add_subplot(133)
d3plot.plot(ax = ax1, linewidth=0.5, edgecolor='0.4', column='mean', cmap='Greens', legend=True, scheme='quantiles')
ax1.set_title('Average number of followers', fontdict={'fontsize': 20}, loc='left')
ax1.get_legend().set_bbox_to_anchor((.4, .3))
ax1.axis('off')


plt.savefig('fig1.png')
plt.show()
```
